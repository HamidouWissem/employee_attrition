## Introduction

Employee Attrition is is one of the foremost concerns of Human Resources (HR) in firms: it refers to the gradual loss of employees over time. Since people are companies’ most important assets, high employee turnover (or attrition) can have several impacts on their organization and their intern strategy. Consequently, HR professionals often try to design company compensation programs, work culture and motivation systems that help the organization retain top employees. To design such a policy, it can be useful to identify exactly which individuals might leave based on patterns that might be present in existing data.


<h3> Business problem overview </h3>

Not all employees are equally performant, some employees with specific knowledge, large experience and particular skills are more valuable than other employees. 

Bill Gates said once about Microsoft : “ You take away our top 20 employees and we become a mediocre company. ” 


When top-performing employees who are responsible for driving sales and increasing revenue become demotivated and start looking for the exit, this is known as negative attrition, and it is an implication of a larger, more serious problem within an organization. Many studies show that the total cost of losing an employee can range from tens of thousands of dollars to 1.5-2X annual salary.
Losing an employee has a cost and is a major problem for many reasons :

1. ** Loss of an employee costs money :**
Employee churn rates can be as high as 12–15% annually.

2. ** Loss of an employee adversely affects ongoing projects and services, which leads to dissatisfaction among customers and other stakeholders. **

3. ** Lost engagement and cultural impact : **
It affects ongoing work and productivity of existing employees who see high turnover disengage and lose productivity. Moreover, whenever someone leaves others take time to ask "why?"

4. ** Lost productivity :**
It takes time and efforts for new employees to achieve the same levels of expertise and productivity. 

5. ** Cost of hiring a new person : **
It is difficult to find suitable replacements for employees, particularly those with high experience and special skills. Furthermore, it takes time, effort and money to recruit new employees via advertising, interviewing, screening and hiring. 

6. ** Cost of onboarding a new person : **
Training and management time cost for the company.

<h3> Estimation of the cost of turnover  </h3>


We have seen that we can associate a real financial value to employee churn. We can estimate the cost of employee attrition to show the concrete impact of this phenomenon on a company. 
We quantify the company’s cost  of employee turnover as equal to the number of regrettable departures times the average cost of those departures. 
Since it is difficult to take into account intangible costs like employee morale, we focus on the following four parameters :
- Cost of hiring
- Cost of onboarding and training
- Cost of learning and development (to reach the maximal productivity)
- Cost of time with unfilled role (time to look for someone else)
So we can now describe the overall annual cost of turnover to be:

$$(Hiring + Onboarding + Development + Unfilled Time)
*  (Annual Number of employees * Annual Turnover Percentage) $$

As an example, if you are a $150$ person company with $11 \%$ annual turnover, and you spend $\$25k$ per person on hiring, $\$10k$ on each of turnover and development, and lose $\$ 50k$ of productivity opportunity cost on average when refilling a role, then your annual cost of turnover would be about $1.57$ million.
Reducing this by just $20\%$, for example, would immediately yield over $\$300k$ in value. And that says nothing of the emotional impact of losing great people.

** What's the cost of turnover in companies ?**

According to the Society for Human Resource Management’s (SHRM’s) new Human Capital Benchmarking ReportThe average cost-per-hire is $\$4,129$, while the average time it takes to fill a given position is $42$ days. In terms of employee retention, the average employee tenure is eight years, the annual turnover rate is $19\%$t and the involuntary turnover rate is $8\%$. These statistics remained stable when compared to last year’s findings. The report analyzes data collected from the SHRM Human Capital Benchmarking Survey in $2016$. The survey of $2,048$ respondents measured employment, retirement, compensation, employer-paid tuition and other people metrics. 

<h3>  Prediction task  </h3>


In this business case, our goal is to understand the key variables that influence employee turnover and to predict if a given employee is going to leave or not the organization. In fact, predictive models for employee churn aim to accurately identify the attrition of “valuable” employees which would help the firm understand the major causes of the phenomenon in order to plan new retention and recruitment strategies within the company. 

Therefore, predicting attrition aims to address a real-life question for any organization : How can we identify which employees are likely to quit ? 

Hence a good prediction can  improve the Turnover Rate which is a KPI especially for the HR strategy.

### Prediction use

Accurate employee turnover predictions enable organisations to take action for retention or succession planning of employees. It can lead to work on the most important factors explaining attrition to reduce this rate. These predictions can allow the company to target some people which are likely to leave the company and offer them advancement and developmental opportunities for capable and ambitious workers to take advantage of. Furthermore, it gives hints on which parameters to improve in the company’s strategy: improve the management, the hiring process, the training, or look for more accurate job profiles ? It can be better to invest a little in employee training and recognition than have to replace employees.


Our goal is to classify each employee into one of these categories.

### Metric definition 
- - - 

** - Use of MCC score: ** <br>
- The Matthews correlation coefficient is used as a measure of the quality of classification since it takes into account true and false positives and negatives and is generally regarded as a balanced measure which can be used even if the classes are of very different sizes (which is suitable for our case since classes have not the same size: classes 2,3,4 are smaller than the other classes).<br>
- The MCC is a correlation coefficient between the observed and predicted binary classifications: it returns a value between −1 and +1. <br>
A coefficient of +1 represents a perfect prediction, 0 no better than random prediction and −1 indicates total disagreement between prediction and observation.<br>
- Other measures like the accuracy, are not useful when classes are of very different sizes. For example, assigning every object to the larger set achieves a high proportion of correct predictions, but is not generally a useful classification. That's why we should not use accuracy as a reference metric for our case but rather observe  the MCC score.
<br>
- The MCC can be calculated directly from the confusion matrix using the formula:
$${\displaystyle {\text{MCC}}={\frac {\sum _{k}\sum _{l}\sum _{m}C_{kk}C_{lm}-C_{kl}C_{mk}}{{\sqrt {\sum _{k}(\sum _{l}C_{kl})(\sum _{k'|k'\neq k}\sum _{l'}C_{k'l'})}}{\sqrt {\sum _{k}(\sum _{l}C_{lk})(\sum _{k'|k'\neq k}\sum _{l'}C_{l'k'})}}}}}$$

** - Use of Cohen-Kappa Metric: ** <br>
- The Kappa statistic (or value) is a metric that compares an Observed Accuracy with an Expected Accuracy (random chance).
- It takes into account random chance, which generally means it is less misleading than simply using accuracy as a metric (an Observed Accuracy of 80% is a lot less impressive with an Expected Accuracy of 75% versus an Expected Accuracy of 50%)
- Computation of Observed Accuracy and Expected Accuracy is integral to comprehension of the kappa statistic, and is most easily illustrated through use of a confusion matrix.
- Kappa is always less than or equal to 1. A value of 1 implies perfect agreement and values less than 1 imply less than perfect agreement

### Feature extractor

The feature extractor implements a `transform` member function. It is saved in the file [`submissions/starting_kit/feature_extractor.py`](/edit/submissions/starting_kit/feature_extractor.py). It receives the pandas dataframe `X_df` defined at the beginning of the notebook. It should produce a numpy array representing the extracted features, which will then be used for the classification.  

Note that the following code cells are *not* executed in the notebook. The notebook saves their contents in the file specified in the first line of the cell, so you can edit your submission before running the local test below and submitting it at the RAMP site.

````python
%%file submissions/starting_kit/feature_extractor.py
# -*- coding: utf-8 -*-

import pandas as pd
import numpy as np 


def add_missing_dummies( X_df, cols ):
    missing_dums = set( cols ) - set( X_df.columns )
    for i in missing_dums:
        X_df[i] = 0

def final_dummies_fix( X_df, cols ):  

    add_missing_dummies( X_df, cols )
# Make sure all columns exists after the add of missing dummies
    try:
        assert( set( cols ) - set( X_df.columns ) == set())
    except AssertionError:
    	print("you haven't added all the missing dummies !!!")

    #print ( "perfect add !!")
# Here we check if the  columns in the test and the train samples are the same.
    ext_columns = set( X_df.columns ) - set( cols )
    if ext_columns:
        print ("there are extra columns in your features:", ext_columns)
# Organizing features and make sure that the order of data in the train is always the same 
    X_df = X_df[ cols ]
    return X_df



def dummify( df):
	""" This function is used to generate dummy variables from some features """
	# dummy of department
	dum_dep = pd.get_dummies(df['department'])
	df = df.drop('name', axis=1)
	df = df.drop('department', axis=1)
	df = df.join(dum_dep)
	
	# dummy of salary
	dum_sal = pd.get_dummies(df['salary'])
	df = df.drop('salary', axis=1)
	df = df.join(dum_sal)
	return (df)



class FeatureExtractor():

    def __init__(self):
        self.columns=['Unnamed: 0', 'satisfaction_level', 'last_evaluation','number_projects', \
                      'average_monthly_hours', 'time_spent_company','work_accident', 'promotion_last_5_years', \
                      'salary_level', 'IT','RandD', 'accounting', 'hr', 'management', 'marketing', 'product_mng','sales',\
                      'support', 'technical', 'high', 'low', 'medium']

    def fit(self, X_df, y=None):
    	return X_df

    def fit_transform(self, X_df, y=None):
        return self.transform(X_df)

    def transform(self, X_df):
        X = dummify(X_df)
        return final_dummies_fix(X,self.columns)
````

### Classifier

The classifier follows a classical scikit-learn classifier template. It should be saved in the file [`submissions/starting_kit/classifier.py`](/edit/submissions/starting_kit/classifier.py). In its simplest form it takes a scikit-learn pipeline, assigns it to `self.clf` in `__init__`, then calls its `fit` and `predict_proba` functions in the corresponding member funtions.

````python
%%file submissions/starting_kit/classifier.py
# -*- coding: utf-8 -*-

from sklearn.base import BaseEstimator
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

class Classifier(BaseEstimator):
    def __init__(self):
        #self.clf = RandomForestClassifier()
        self.clf = LogisticRegression()

    def fit(self, X, y):
        #self.clf.fit(X.todense(), y)
        self.clf.fit(X, y)

    def predict(self, X):
        return self.clf.predict(X)

    def predict_proba(self, X):
        return self.clf.predict_proba(X)
````
